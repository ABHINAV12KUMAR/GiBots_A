```python
# Remove Duplicates from train_data and also remove the corresponding rows in train_labels as well
original_indices = train_data.index
train_data = train_data.drop_duplicates()
dropped_indices = original_indices.difference(train_data.index)
print("Total rows dropped : " + str(len(dropped_indices)))
train_labels = train_labels.drop(dropped_indices)
```

```python
train_data.corr
```

```python
# Trying to find the columns that are categorical and needs to be encoded
encode_columns = []
for index,element in enumerate(train_data.iloc[0,:]):
  if element == 'YES' or element == 'NO':
    encode_columns.append("x" + str(index+1))
    # print(index+1)
print(encode_columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,f1_score,accuracy_score,hamming_loss,precision_score,recall_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
# Feature Selection
train_data = train_data.drop(columns=["x3","x4","x34","x35","x61","x64","x65","x91","x94","x95"])
```

```python
train_data
```

```python
# Applying train_test split
X_train, X_val, y_train, y_val = train_test_split(train_data, train_labels, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
# Handling Categorical values "YES" or "NO"
from sklearn.compose import ColumnTransformer

non_encode_cols = [col for col in X_train.columns if col not in encode_columns]

transformer = ColumnTransformer(transformers = [
    ('tnf1', OrdinalEncoder(), encode_columns)
],remainder='passthrough')

encode_data1 = transformer.fit_transform(X_train)
encode_data2 = transformer.transform(X_val)

print(encode_data1)

transformed_xtrain = pd.DataFrame(encode_data1, columns= encode_columns + non_encode_cols)
transformed_xval = pd.DataFrame(encode_data2, columns= encode_columns + non_encode_cols)
```

```python
transformed_xtrain
```

```python
# Missing Value Imputation

transformer2 = ColumnTransformer(transformers = [
    ('tnf2', SimpleImputer(strategy="most_frequent"), encode_columns)
],remainder='passthrough')

impute_data1 = transformer2.fit_transform(transformed_xtrain)
impute_data2 = transformer2.transform(transformed_xval)

print(impute_data1)

transformed_xtrain = pd.DataFrame(impute_data1, columns= encode_columns + non_encode_cols)
transformed_xval = pd.DataFrame(impute_data2, columns= encode_columns + non_encode_cols)
```

```python
# Trying the accuracy of all the models and chosing the best model
def try_models(x_train,y_train,x_test,y_test,model_name = RandomForestClassifier(n_estimators=100 , random_state=42),multi_label_model = LabelPowerset):

  warnings.filterwarnings("ignore")
  model =  multi_label_model(model_name)
  print(model.fit(x_train, y_train))
  y_pred = model.predict(x_test)

  if multi_label_model == MultiOutputClassifier:
    # Because MultiOutputClassifier model has no metrics hence we have to create them ourselves
    y_val_np = np.array(y_val)
    y_pred_np = np.array(y_pred)

    # Calculate Hamming Loss
    hamming_loss_value = np.mean(np.not_equal(y_val_np, y_pred_np))
    print("Hamming Loss is : " + str(hamming_loss_value))

    # Calculating Accuracy
    temp = 0
    for i in range(y_val.values.shape[0]):
        temp += sum(np.logical_and(y_val.values[i], y_pred[i])) / sum(np.logical_or(y_val.values[i], y_pred[i]))
    print("Accuracy is : " + str(temp / y_val.values.shape[0]))

    # Calculating 0-1 Loss
    loss01 = np.any(y_val != y_pred, axis=1).mean()
    print("0-1 Loss is : " + str(loss01))

    # Calculating Recall
    temp = 0
    for i in range(y_val.values.shape[0]):
        if sum(y_val.values[i]) == 0:
            continue
        temp+= sum(np.logical_and(y_val.values[i], y_pred[i]))/ sum(y_val.values[i])
    print("Recall is : " + str(temp/ y_val.values.shape[0]))

    # Calculating Precision
    temp = 0
    for i in range(y_val.values.shape[0]):
        if sum(y_pred[i]) == 0:
            continue
        temp+= sum(np.logical_and(y_val.values[i], y_pred[i]))/ sum(y_pred[i])
    print("Precision is : " + str(temp/ y_val.values.shape[0]))

    return model

  print("Accuracy is : " + str(accuracy_score(y_val,y_pred)))
  print("Hamming Loss is : " + str(hamming_loss(y_val,y_pred)))
  print("Recall is : " + str(recall_score(y_val,y_pred,average='weighted')))
  print("Precision is : " + str(precision_score(y_val,y_pred,average = 'weighted')))
  print("F1 Score is : " + str(f1_score(y_val,y_pred,average = 'weighted')))
  warnings.resetwarnings()

  return model

```

## **The Best accuracy is achieved by multi_label_models = LabelPowerset and model_name = RandomForestClassifier **

```python
# The possible multi_label_models are : LabelPowerset,BinaryRelevance,ClassifierChain,MultiOutputClassifier
# Suggested model_name are : MLPClassifier,DecisionTreeClassifier,RandomForestClassifier

model = try_models(x_train=transformed_xtrain,x_test=transformed_xval,y_train= y_train,y_test= y_val)

# After applying GridSearch CV , the best model combinations are the default values i.e. multi_label_models = LabelPowerset , model_name = RandomForestClassifier
```

```python
# Function that will provide us with the output for the given input
def get_output(model , test_data):

  test_data.columns = traindata_columns.insert(0,"id")
  test_data = test_data.drop([0])
  # Feature Selection of the input
  test_data = test_data.drop(columns=["x3","x4","x34","x35","x61","x64","x65","x91","x94","x95"])
  # Encoding Categorical data
  encode_data3 = transformer.transform(test_data)
  transformed_xtest = pd.DataFrame(encode_data3, columns= encode_columns + non_encode_cols)
  # Missing value imputation of input
  impute_data4 = transformer2.transform(transformed_xtest)
  transformed_xtest = pd.DataFrame(impute_data4, columns= encode_columns + non_encode_cols)
  # Model Prediction
  output = model.predict(transformed_xtest)
  return output.toarray()
```

```python
answer = get_output(model,test_data.copy())
```

```python
# Changing the output generated in the form that is mentioned in the samplesubmission_small.csv
output_dataset = pd.DataFrame(columns = ["id_label","pred"])

for ind1,element in enumerate(answer):
  for ind2,num in enumerate(element):
    new_data = {
        "id_label" : str(test_data.iloc[ind1,0]) + "_y" + str(ind2+1),
        "pred" : num
    }
    # Convert the new data to a DataFrame
    new_row = pd.DataFrame([new_data])

    # Concatenate the existing DataFrame with the new row
    output_dataset = pd.concat([output_dataset, new_row], ignore_index=True)
```
